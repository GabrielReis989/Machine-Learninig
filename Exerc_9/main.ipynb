```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

```python
dados = pd.read_csv('C:\\Users\\reisg\\OneDrive\\Área de Trabalho\\Aquivos_importantes\\Treinamento\\Exerc_9\\Data_train_reduced.csv')
dados.head() 
```

```python
dados.drop(['Respondent.ID', 'Product.ID', 'Product'], axis=1, inplace=True)
```

```python
dados.isnull().sum()
```

```python
dados['Instant.Liking'].fillna(dados['Instant.Liking'].mean(), inplace=True)
dados['q8.2'].fillna(dados['q8.2'].mean(), inplace=True)
dados['q8.7'].fillna(dados['q8.7'].mean(), inplace=True)
dados['q8.8'].fillna(dados['q8.8'].mean(), inplace=True)
dados['q8.9'].fillna(dados['q8.9'].mean(), inplace=True)
dados['q8.10'].fillna(dados['q8.10'].mean(), inplace=True)
dados['q8.12'].fillna(dados['q8.12'].mean(), inplace=True)
dados['q8.17'].fillna(dados['q8.17'].mean(), inplace=True)
dados['q8.18'].fillna(dados['q8.18'].mean(), inplace=True)
dados['q8.19'].fillna(dados['q8.18'].mean(), inplace=True)
dados['q8.20'].fillna(dados['q8.18'].mean(), inplace=True)
```

```python
dados.isnull().sum()
```

```python
y = dados['Instant.Liking']
x = dados.drop('Instant.Liking', axis=1) 
```

```python
kfold = StratifiedKFold(n_splits=10)
modelo = LogisticRegression(penalty='l2',max_iter=500).fit(x,y)
resultado = cross_val_score(modelo,x,y,cv=kfold)  
```

```python
print(resultado.mean())  
```

```python
valores_C = np.array([0.01,0.02,21,23,25])
regularizacao = ['l2']
grid = {'C':valores_C, 'penalty':regularizacao}
```

```python
gridSeachcv = GridSearchCV(estimator=modelo, param_grid=grid, cv=10)
gridSeachcv.fit(x,y) 
```

```python
print(f'{gridSeachcv.best_estimator_.C} | {gridSeachcv.best_estimator_.penalty} | {gridSeachcv.best_score_}')
```
