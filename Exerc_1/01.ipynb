```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
x,y =make_regression(n_samples=200, n_features=1, noise=30)

plt.scatter(x,y) 
```

```python
modelo = LinearRegression()
modelo.fit(x,y)
```

```python
print(f'Coeficiente Angular: {modelo.coef_} e Coeficiente Linear: {modelo.intercept_}')
```

```python
plt.scatter(x,y) 
distancia = np.arange(-6,6,1)
plt.plot(distancia, (modelo.coef_)*distancia - modelo.intercept_, color = 'yellow')

plt.show() 
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.30)
```

```python
modelo.fit(x_treino, y_treino)

print(modelo.score(x_teste, y_teste)*100)
```
