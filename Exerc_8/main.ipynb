```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import cross_val_score
pd.set_option('Display.max_columns', 230)
pd.set_option('display.width', 45)
```

```python
dados = pd.read_csv('C:\\Users\\reisg\\OneDrive\\√Årea de Trabalho\\Aquivos_importantes\\Treinamento\\Exerc_8\\Admission_Predict.csv')
dados.head()
```

```python
dados.drop('Serial No.', axis= 1, inplace=True)
```

```python
dados.head()
```

```python
y = dados['Chance of Admit ']
x = dados.drop('Chance of Admit ', axis=1)
```

```python
kfold = KFold(n_splits=10)
modelo = ElasticNet()
modelo_02 = LinearRegression()
modelo_03 = Ridge()
modelo_04 = Lasso()
```

```python
cross_val_score(modelo_02,x,y, cv=kfold).mean() 
```

```python
valores = {'alpha':[0.26,0.450,0.0025], 'l1_ratio':[0.01,0.005,0.241]}
resultado = RandomizedSearchCV(estimator=modelo,param_distributions=valores,n_iter=9, cv=kfold)
resultado.fit(x,y)
```

```python
print(f'Melhor alpha: {resultado.best_estimator_.alpha} | Melhor l1_ratio: {resultado.best_estimator_.l1_ratio} | Melhor score: {resultado.best_score_}')
```

```python
resultado_02 = GridSearchCV(estimator=modelo, param_grid=valores,cv=kfold) 
resultado_02.fit(x,y) 
```

```python
print(f'Melhor alpha: {resultado_02.best_estimator_.alpha} | Melhor l1_ratio: {resultado_02.best_estimator_.l1_ratio} | Melhor score: {resultado_02.best_score_}')
```
